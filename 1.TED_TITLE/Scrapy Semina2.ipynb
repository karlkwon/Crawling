### 2. Defining the Items you will extract

Items are containers that will be loaded with the scraped data; <b>they work like simple Python dicts</b>. While you can use plain Python dicts with Scrapy, Items provide additional protection against populating undeclared fields, preventing typos.

### 3. Writing a spider to crawl a site and extract Items

Spiders are classes that you define and Scrapy uses to scrape information from a domain (or group of domains).<br>

name, start_urls, parse()

#### Crawling


<u><i>What just happened under the hood?</i></u>

Scrapy creates scrapy.Request objects for each URL in the start_urls attribute of the Spider, and assigns them the parse method of the spider as their callback function.

These Requests are scheduled, then executed, and scrapy.http.Response objects are returned and then fed back to the spider, through the parse() method.


```python
!scrapy crawl dmoz
```

### Youtube subtitle download

```python
# http://stackoverflow.com/questions/89228/calling-an-external-command-in-python
import os
os.system("pws")
```

```python
import commands
commands.getoutput("pwd")
```

```python
# youtube-dl https://www.youtube.com/watch?v=e7S8jWh6AEs --all-subs --skip-download -o '/data/%(title)s.%(ext)s'
# options = "--all-subs --skip-download -o '/data/%(title)s.%(ext)s'"

def downloadYoutubeSubtitle(_id):
    url = 'https://www.youtube.com/watch?v=' + _id
    options = "--all-subs --skip-download -o '/data/" + _id + ".%(ext)s'"
    cmd = 'youtube-dl ' + url + ' ' + options

    commands.getoutput(cmd)
```

### Get Youtube list

```python
import math
```

```python
from pprint import pprint
from datetime import datetime
from elasticsearch import Elasticsearch
es = Elasticsearch(['localhost:9200'],
                  use_ssl=False)
```

```python
cnt = es.count(index='ted_test', doc_type='items')
cnt = cnt['count']
print (cnt)
```

```python
loopCnt = int(math.ceil(cnt / 20.))

for i in range(loopCnt):
    res = es.search(index='ted_test',  body={"from": i*20, "size": 20}, filter_path=['hits.hits._id', 'hits.hits._source'])
    #pprint(res['hits']['hits'])
    print (str(i) + " : " + str(map(lambda x: x['_source']['youtube'], res['hits']['hits'])) )
```

## Download all subtitle

```python
import time

loopCnt = int(math.ceil(cnt / 20.))

_cnt = 0
for i in range(loopCnt):
    res = es.search(index='ted_test',  body={"from": i*20, "size": 20}, filter_path=['hits.hits._id', 'hits.hits._source'])
    ids = map(lambda x: x['_source']['youtube'], res['hits']['hits'])
    
    for _id in ids:
        _cnt += 1
        
        try:
            downloadYoutubeSubtitle(_id)
            print (str(_cnt) + ' Downloaded: ' + _id)
        except:
            print ('Error: ' + _id)
            
        time.sleep(2)
```
